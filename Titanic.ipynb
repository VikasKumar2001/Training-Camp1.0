```python
def agge(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 38
    elif Pclass==2:
      return 30
    elif Pclass==3:
      return 28
  else:
    return Age 

       
```

```python
titanic['Age']=titanic[['Age','Pclass']].apply(agge,axis=1)
```
