```python
import seaborn as sns 
%matplotlib inline
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

**DISPLOT**

```python
sns.displot(data=df['total_bill'],bins=100)
```

```python
sns.displot(data=df['total_bill'],bins=100)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
sns.rugplot(df['total_bill'])
```

**BAR PLOT AND COUNT PLOT**


```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

**HEAT MAP **

```python
tp=df.pivot_table(values='total_bill',index='sex',columns='size')
```

```python
sns.heatmap(tp,cmap='coolwarm',annot=True,linewidth=1,linecolor='red')
```

```python
X2 = df.corr()
X2
```

```python
sns.heatmap(X2,annot=True) #sns.heatmap(df.corr())
```
