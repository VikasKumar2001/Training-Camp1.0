```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("ParisHousing.csv")
data
```

```python
data.isnull().sum()      #no null values are present
```

```python
#can be viewed using
sns.heatmap(data.isnull(),yticklabels=False,cbar=False)
```

```python
sns.countplot(x='hasYard',data=data)
```

```python
sns.countplot(x = 'hasPool',data=data)
```

```python
plt.figure(figsize = (16, 12))
ax = sns.countplot(x = 'made',data=data)
ax
```

```python
sns.displot(x='floors',hue='hasPool',data=data)
```

```python
from matplotlib import pyplot as plt
plt.scatter(data['price'], data['squareMeters'],alpha = 0.1)
plt.xlabel('Price')
plt.ylabel('Square Meters')
plt.show()
```

```python
plt.scatter(data['price'], data['made'], alpha = 0.1)
plt.xlabel('Price')
plt.ylabel('Year Made')
plt.show()
```

```python
corr = data.corr()
plt.figure(figsize = (16, 12))
sns.heatmap(corr, annot = True , cmap="magma")
plt.show()
```
